```python
import numpy as np
import pandas as pd

import sklearn
```

```python
n1 = np.random.normal(loc=0.0, scale=1.0, size=2000)
n2 = np.random.normal(loc=0.0, scale=1.0, size=2000)

df = pd.DataFrame(data={
    'n1':n1,
    'n2':n2,
})
```

```python
df.plot.scatter('n1','n2')
```

```python
(df['n1']+df['n2']).plot.hist()
```

```python
(df['n1']-df['n2']).plot.hist()
```

```python
from sklearn.metrics import accuracy_score,mean_squared_error
mean_squared_error(df['n1'],df['n2'])
```

```python
df['n1'] = (df['n1']*100).astype(int)
df['n2'] = (df['n2']*100).astype(int)
```
